```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2
import numpy as np  
```

```python

sample = Image.open("../img1.jpg")
plt.imshow(sample)
ax = plt.gca()
rect1 = patches.Rectangle((410,245),
                 90,
                 240,
                 linewidth=3,
                 color='cyan',
                 fill = True,
                 alpha = .45)
rect2 = patches.Rectangle((5,310),
                 145,
                 90,
                 linewidth=3,
                 color='purple',
                 fill = True,
                 alpha = .15)

ax.add_patch(rect1)
ax.add_patch(rect2)


img = cv2.imread('../img1.jpg',0)
edges = cv2.Canny(img,260,350)   # canny edge detector

img1 = cv2.merge((img,img,img))  # creat RGB image from grayscale

img1[edges == 255] = [255, 0, 0] # turn edges to red
img2=img1.copy()


img1 = ax.imshow(img1)
img1.set_clip_path(rect1)
img2 = ax.imshow(img2)
img2.set_clip_path(rect2)
bbox_props = dict(boxstyle="round", fc="w", ec="0.5", alpha=0.9)
ax.text(70, 270, "obj 1, cls 81", ha="center", va="center", size=7,
        bbox=bbox_props)
ax.text(460, 230, "obj 2, cls 70", ha="center", va="center", size=7,
        bbox=bbox_props)
x = [460, 430, 435,440,420,410,430,415,420,410,410,500,500,490,493,480,500,490,470,475,480]
y = [480, 470, 430,410,400,380,350,310,300,290,245,245,290,300,310,350,380,400,410,430,470]

x2 = [5,50,100,140,145,100,50,10,5]
y2= [330,310,320,330,350,380,385,380,375]


ax.fill(x, y, color = 'cyan',fill=False)
ax.fill(x2, y2, color = 'pink',fill=False)
plt.show()
```
