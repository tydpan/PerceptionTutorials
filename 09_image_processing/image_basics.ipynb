**IMAGE BASICS **

Image in computer vision can be viewed as a container of pixels, where each pixel have values representing the intensity of the light captured.

The Image can be 2 dimensional (height, width) like a 2D array or 3 dimensional (height, width, channel) like a 3D channel. 

For Example : A 2D image of (1920,1080) means it is a 2D array of shape 1920-1080, and it contains 1920*1080 pixels with intensity values. These intensity values can be of byte, unsigned/signed int, float, double type.

```python
## in this code we will create and display a simple grayscale image 

import numpy as np 
import matplotlib.pyplot as plt

# define the image shape 
image_shape = (200, 200)

# create the image with all pixel values as 1

image_2d = np.ones(image_shape)

# lets plot this basic image
plt.imshow(image_2d, cmap='gray')
plt.show()

## YOUR TASK 
#### Try to change this simple "image_2d" such that it has a (100,100) box in the middle
#### The box in the middle should have the maximum possible values of the pixels 
#### you can assume that each pixel is 8-bit unsigned integer 
## HINT: use matrix slicing

# your code here 

# plot the updated image 
plt.imshow(image_2d,cmap='gray')
plt.show()
```

Now that we have a basic understanding of 2D images, lets move onto 3D images.

3D images are 3D matrices of the form (height, width, channel). This 3D dimension called channel is what encodes the color information of the images we see. As you would have noted above the image doesn't contain any color information (being grayscale) 

color images can be represented in many forms like RGB, HSV, and HSL these are also called as color models. 

Here we will work with RGB model.
If you want to learn more about color spaces then check out this link  
[color spaces](https://programmingdesignsystems.com/color/color-models-and-color-spaces/index.html)

RGB stands for Red-Green-Blue. Why red, green blue ?? 

This is beacuse, it is possible to generate any other color using these 3 colors with different intensities and mixing ratio. 



```python
# in this code block we will generate and display a color image 


# define image shape 
# here the 3rd dimension is the channel of shape 3 (red, green, blue)
image_shape_3d = (200,200,3)


# define the image 
image_3d = np.zeros(image_shape_3d)


# plot the color image 
plt.imshow(image_3d)
plt.show()
```

```python
# as you can see this image does not have any color information 
# lets see this same image with different colors 

# red image 
red_image = image_3d*1
red_image[:,:,0] = 255

plt.imshow(red_image)
plt.show()

# blue image 
blue_image = image_3d*1
blue_image[:,:,2] = 255

plt.imshow(blue_image)
plt.show()

# green image 
green_image = image_3d*1
green_image[:,:,1] = 255

plt.imshow(green_image)
plt.show()
```
